```python
##class VotingSystem :
  
    def __init__(self):
        self.candidates = {
            "1": {"name": "John Doe", "votes": 0},
            "2": {"name": "Jane Smith", "votes": 0},
            "3": {"name": "Bob Johnson", "votes": 0},
        }

    def display_menu(self):
        print("\nVoting System Menu:")
        print("1. Cast Vote")
        print("2. View Candidates")
        print("3. View Results")
        print("4. Exit")

    def display_candidates(self):
        print("\nCandidates:")
        for id, candidate in self.candidates.items():
            print(f"{id}. {candidate['name']}")

    def cast_vote(self):
        self.display_candidates()
        choice = input("\nEnter candidate number to vote: ")
        if choice in self.candidates:
            self.candidates[choice]["votes"] += 1
            print("Vote cast successfully!")
        else:
            print("Invalid candidate number.")

    def view_results(self):
        print("\nVoting Results:")
        for id, candidate in self.candidates.items():
            print(f"{candidate['name']}: {candidate['votes']} votes")

    def run(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")
            if choice == "1":
                self.cast_vote()
            elif choice == "2":
                self.display_candidates()
            elif choice == "3":
                self.view_results()
            elif choice == "4":
                print("Exiting Voting System. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    voting_system = VotingSystem()
    voting_system.run()
    
    
            
    
    



    
        
        

    

    
    
            
            
  
```
